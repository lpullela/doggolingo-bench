This script will combine the sentence examples pulled from the utterances with those pulled from the titles into one csv file. 

```python
import pandas as pd
import numpy as np
import json
```

```python
with open('sents_from_utterances.json', 'r') as file:
    ut_data = json.load(file)

with open('sentence_examples.csv', 'r') as file:
    sent_data = pd.read_csv(file, delimiter=',', on_bad_lines='warn')
```

```python
utter_df = pd.DataFrame(ut_data)
utter_df = utter_df.T
utter_df.head()
```

```python
utter_df.reset_index(drop=True, inplace=True)
```

```python
word_counts = utter_df["word"].value_counts()
word_counts = word_counts[:100]
print(word_counts.index.tolist())
```

```python
utter_df = utter_df[['word', 'text']]
sent_data = sent_data.rename(columns={'word':'word', 'sentence': 'text'})
```

```python
new_df = pd.concat([utter_df, sent_data])
new_df.drop_duplicates(inplace=True)
new_df = new_df.sort_values(by='word')
new_df = new_df[new_df.word != "dogs"]
new_df = new_df[new_df.word != "petsmart"]
new_df = new_df[new_df.word != "clickertraining"]
new_df = new_df[new_df.word != "rehoming"]
new_df = new_df[new_df.word != "chihuahuas"]
new_df = new_df[new_df.word != "playdates"]
new_df = new_df[new_df.word != "offleash"]
new_df = new_df[new_df.word != "catsthegame"]
new_df = new_df[new_df.word != "dogfoodadvisor"]
new_df = new_df[new_df.word != "catssittingdown"]
new_df = new_df[new_df.word != "dogfoodadvisor"]
new_df = new_df[new_df.word != "litterbox"]
new_df = new_df[new_df.word != "rehome"]
new_df = new_df[new_df.word != "shiba"]
new_df = new_df[new_df.word != "littermate"]
new_df = new_df[new_df.word != "labradoodle"]
new_df = new_df[new_df.word != "vizsla"]
new_df = new_df[new_df.word != "housetrain"]
new_df = new_df[new_df.word != "bichon"]
new_df = new_df[new_df.word != "counterconditioning"]
new_df.to_json('words_with_text.json', orient='records', lines=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

word_counts = new_df["word"].value_counts()
freq_counts = word_counts.value_counts()
freq_df = freq_counts.rename_axis("value_count").reset_index(name="number_of_items")
freq_df.columns = ['value_count', 'number_of_items']

sns.set(font_scale=0.8)

plt.figure(figsize=(25, 6))
sns.barplot(data=freq_df, x = 'value_count', y = "number_of_items")
plt.xlabel("Number of Sentences", fontsize=8)
plt.ylabel("Number of Words")
plt.xticks(rotation=90)
```

```python
# Print initial sizes
print(f"Initial utter_df size: {len(utter_df)}")
print(f"Initial sent_data size: {len(sent_data)}")

# After concatenation
new_df = pd.concat([utter_df, sent_data])
print(f"After concatenation size: {len(new_df)}")

# After dropping duplicates
new_df.drop_duplicates(inplace=True)
print(f"After dropping duplicates size: {len(new_df)}")

# Count before filtering
print(f"\nBefore filtering - total rows: {len(new_df)}")

# Create a list of words to filter
words_to_filter = ["dogs", "petsmart", "clickertraining", "rehoming", "chihuahuas",
                  "playdates", "offleash", "catsthegame", "dogfoodadvisor",
                  "catssittingdown", "litterbox", "rehome", "shiba", "littermate",
                  "labradoodle", "vizsla", "housetrain", "bichon", "counterconditioning"]

# Count occurrences of each word being filtered
for word in words_to_filter:
    count = len(new_df[new_df.word == word])
    if count > 0:
        print(f"Found {count} rows with word '{word}'")

# More efficient filtering
new_df = new_df[~new_df.word.isin(words_to_filter)]

# Final count
print(f"\nAfter all filtering - final size: {len(new_df)}")

# Check value counts of words
print("\nTop 10 most frequent words:")
print(new_df['word'].value_counts().head(10))

new_df.to_json('words_with_text.json', orient='records', lines=True)
```
